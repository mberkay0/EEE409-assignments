```python
import turicreate
```

```python
from google.colab import drive
drive.mount('/content/gdrive', force_remount=True)

import warnings
warnings.filterwarnings("ignore")
```

```python
data = turicreate.SFrame("/content/gdrive/MyDrive/Introduction to Machine Learning/datas/people_wiki.sframe")
```

```python
data[data['name']=='Victoria Beckham']
```

```python
turicreate.SFrame.export_csv(data, 'people_wiki.csv')
```

```python
data['word_count']=turicreate.text_analytics.count_words(data['text'])
```

```python
data.head()
```

```python
data['tf-idf']=turicreate.text_analytics.tf_idf(data['word_count'])
```

```python
data.head()
```

```python
Elton_John=data[data['name']=='Elton John']
```

```python
Elton_John.stack('word_count',new_column_name=['word','count'])[['word','count']].sort('count',ascending=False)
```

```python
Elton_John.stack('tf-idf',new_column_name=['word','tf-idf_score'])[['word','tf-idf_score']].sort('tf-idf_score',ascending=False)
```

```python
Victoria_Beckham=data[data['name']=='Victoria Beckham']
Paul_McCartney=data[data['name']=='Paul McCartney']
```

```python
turicreate.distances.cosine(Elton_John['tf-idf'][0],Victoria_Beckham['tf-idf'][0])
```

```python
turicreate.distances.cosine(Elton_John['tf-idf'][0],Paul_McCartney['tf-idf'][0])
```

```python
recommend_docs=turicreate.nearest_neighbors.create(data,label='name',features=['tf-idf'])
```

```python
recommend_docs.summary()
```

```python
recommend_docs_cosine=turicreate.nearest_neighbors.create(data,label='name',features=['tf-idf'],distance='cosine')
```

```python
recommend_docs_cosine.summary()
```

```python
recommend_docs.query(Victoria_Beckham)
```

```python
recommend_docs_cosine.query(Victoria_Beckham)
```

```python
recommend_docs_wordcount=turicreate.nearest_neighbors.create(data,label='name',features=['word_count'],distance='cosine')
```

```python
recommend_docs_wordcount.query(Victoria_Beckham)
```
